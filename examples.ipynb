```python
G = nx.read_gpickle('test_graphs/07 inter-subtype interactions.pkl')
```

```python
from circos import CircosPlot
```

```python
nodes = G.nodes()
edges = [(u, v) for u, v in G.edges() if 'H3N2' in [u, v] and u != v]

nodeprops = dict(facecolor='red', alpha=0.5)
edgeprops = dict(edgecolor='orange', alpha=0.75, ls='--')

c = CircosPlot(nodes, edges, 10,
               #nodecolor=plt.cm.viridis(np.arange(len(nodes)) / len(nodes)),
               #edgecolor='black',
               #edgeprops=edgeprops,
               #nodeprops=nodeprops
              )
c.draw()
```

```python
c = CircosPlot(nodes, edges, 10,
               nodecolor=plt.cm.viridis(np.arange(len(nodes)) / len(nodes)),
               #edgecolor='black',
               #edgeprops=edgeprops,
               #nodeprops=nodeprops
              )
c.draw()
```
